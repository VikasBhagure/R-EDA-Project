#### table
** used to create  categorical representation of data with variale name and the frequency in the form of a table. 

```r
count<-table(mtcars$mpg)
```

```r
print(count)
```

```r
barplot(count)
```

```r
barplot(mtcars$gear) # not used since it doesnot give proper details
```

```r
barplot(count,col="red",main="Cars",xlab="mpg",ylab="values")
```

```r
colors<-c("red","green","blue","pink","red","green","blue","pink","red")
barplot(count,col=colors,main="Cars",xlab="mpg",ylab="values")
```

```r
barplot(count,col=colors,main="Cars",xlab="mpg",ylab="values",horiz=TRUE)
```

```r
counts<-table(mtcars$vs,mtcars$gear)
barplot(counts,main="CAr Distribution by Gears and VS",xlab="Number of Gears",
col=c("grey","cornflowerblue"),
legend=rownames(counts))
```

```r
counts<-table(mtcars$vs,mtcars$gear)
barplot(counts,main="CAr Distribution",xlab="Number of Gears",
col=c("green","blue"),
legend=rownames(counts),beside=TRUE)
```

```r
counts<-table(mtcars$vs,mtcars$gear,mtcars$cyl)
slices<-c(mtcars$vs,mtcars$gear,mtcars$cyl)
pie(counts,slices,main="Pie Chart")
```

```r
slices<-c(10,12,4,16,8)
pct<-round((slices/sum(slices))*100)
l<-paste(c("US","UK",'austria',"germany","france"),' ',pct,"%",sep='')
pie(slices,labels=l,col=rainbow(5),main="piechart")
```
