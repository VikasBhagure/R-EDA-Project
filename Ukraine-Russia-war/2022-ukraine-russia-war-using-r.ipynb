```r
russia_losses_equipment=read.csv("russia_losses_equipment.csv")
```

```r
russia_losses_personnel=read.csv('russia_losses_personnel.csv')
head(russia_losses_personnel, 3)
```

```r
russia_losses_equipment$date <- as.factor(russia_losses_equipment$date);head(russia_losses_equipment, 3)
```

```r
russia_losses_equipment$date <- as.Date(russia_losses_equipment$date, "%Y-%m-%d");head(russia_losses_equipment, 3)
```

```r
russia_losses_equipment %>% mutate(aircraft_lag1 = lag(aircraft, 1), aircraft_diff1 = c(NA, diff(aircraft, lag = 1)), helicopter_lag1 = lag(helicopter, 1), helicopter_diff1 = c(NA, diff(helicopter, lag = 1)),tank_lag1 = lag(tank, 1), tank_diff1 = c(NA, diff(tank, lag = 1))) -> russia_losses_equipment1
```

```r
russia_losses_equipment1$aircraft_lag1[1] <- 10 
russia_losses_equipment1$aircraft_diff1[1] <- 0
russia_losses_equipment1$helicopter_lag1[1] <- 7 
russia_losses_equipment1$helicopter_diff1[1] <- 0
russia_losses_equipment1$tank_lag1[1] <- 80 
russia_losses_equipment1$tank_diff1[1] <- 0
```

```r
russia_losses_equipment1.xtss <- as.xts(russia_losses_equipment1[, 18:23], order.by = russia_losses_equipment1$date)
```

# <b> <span style='color:#FF5733'>4.</span> Visualize time series data </b>

```r
summ <- russia_losses_equipment1 %>% summarise(mean_aircraft_diff1 = mean(aircraft_diff1))
summ
```

```r
p1 <- ggplot(data = russia_losses_equipment1, aes(x = date, y = aircraft)) + geom_line(aes(group = 1)) + geom_point(shape = 'circle') + labs(title = 'destroyed russian aircraft by date', x = 'Date', y = 'cumulative number') + scale_y_continuous(labels = scales::number_format(big.mark = ',')) + scale_x_date(breaks = '15 day') + theme(axis.text.x=element_text(angle=90,hjust=1))
```

```r
my_color <- c("#7F44F7", "#FDAC44")
```

```r
p2 <- ggplot(data = russia_losses_equipment1, aes(x = date, y = aircraft)) + geom_line(aes(group = 1)) + geom_point(shape = 'circle') + geom_smooth(method = "lm", se = FALSE) + scale_color_manual(values = my_color) + labs(title = 'destroyed russian aircraft by date', x = 'Date', y = 'cumulative number') + scale_y_continuous(labels = scales::number_format(big.mark = ',')) + scale_x_date(breaks = '15 day') + theme(axis.text.x=element_text(angle=90,hjust=1))
```

```r
p3 <- ggplot(data = russia_losses_equipment1, aes(x = date, y = aircraft_diff1)) + geom_line(aes(group = 1)) + geom_point(shape = 'circle') + labs(title = 'destroyed russian aircraft by date', x = 'Date', y = 'number') +scale_y_continuous(labels = scales::number_format(big.mark = ',')) +scale_x_date(breaks = '15 day') + theme(axis.text.x=element_text(angle=90,hjust=1))
```

```r
p4<- ggplot(data = russia_losses_equipment1, aes(x = date, y = aircraft_diff1)) + geom_line(aes(group = 1)) + geom_point(shape = 'circle') + geom_text(aes(label=scales::number(aircraft_diff1, big.mark=',')), size=3.5, vjust=1.5) + geom_hline(aes(yintercept=2.75), colour='red', linetype='dashed', lwd=1) +
  labs(title = 'destroyed russian aircraft by date', x = 'Date', y = 'number') + scale_y_continuous(labels = scales::number_format(big.mark = ',')) + scale_x_date(breaks = '15 day') +theme(axis.text.x=element_text(angle=90,hjust=1))
```

```r
plot_grid(p1,p2,p3,p4,nrow=2,ncol=2)
```

```r
ggplot(data = russia_losses_equipment1, aes(x = date, y = aircraft_diff1)) + geom_point(size = 2) + geom_smooth(method = "lm", se = FALSE, aes(color = "lm")) + labs(title = "destroyed aircraft by date",x = "date", y = "number")
```
